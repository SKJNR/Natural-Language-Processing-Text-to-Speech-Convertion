# Converting Article From Text to Speech

## Problem Description

>From here  main aim is to convert an Article from Text to Speech.For this i'm using an article which published from Newyork Times about NBA players .

```python
#Importing libraries 
from newspaper import Article
import nltk
from gtts import gTTS
import os
```

```python
#Get an article
article=Article('https://www.nytimes.com/2020/08/08/at-home/coronavirus-sports-podcasts.html')
article.download() #Download The Article
article.parse() #Parse The Article
nltk.download('punkt') #Download Package
```

```python
article.nlp() #Apply NLP (Natural Language Processing)
```

```python
#Get The Articles Text And Store It Into A Variable Called 'mytext'
mytext = article.text

#Print The Text
print(mytext)
```

```python
language = 'en' #English
```

```python
#Convert Text To Speech
myobj = gTTS(text=mytext, lang=language, slow=False)
```

```python
from IPython.display import Audio
myobj.save('read_article.mp3')
sound_file = 'read_article.mp3'
Audio(sound_file, autoplay=True)
```
